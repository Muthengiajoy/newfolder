```python
import numpy as np

a = np.array(6)
b = np.array([2,6,8,5])
c = np.array([[2,3,4,5],[2,5,7,9]])
d = np.array([[[3,4,5,6],[2,3,5,7],[2,4,6,8],[1,3,5,6]]])

print("zero dimentional:",a)
print("1-D:",b)
print("2-D:",c)
print("3-D:",d)

print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)


```
